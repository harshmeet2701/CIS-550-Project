```
def removeLeadingAndEndingQuotes(inStr):
  if inStr.startswith('"') and inStr.endswith('"'):
    return inStr[1:-1]
```

```

books_df = pd.DataFrame(columns=['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-S', 'Image-URL-M', 'Image-URL-L'])

temp_list = []
for index in range(0, len(bx_books_df)):
  temp_dict = {}

  temp_dict['ISBN'] = bx_books_df['ISBN;"Book-Title";"Book-Author";"Year-Of-Publication";"Publisher";"Image-URL-S";"Image-URL-M";"Image-URL-L"'][index].split(';')[0]

  temp_dict['Book-Title'] = bx_books_df['ISBN;"Book-Title";"Book-Author";"Year-Of-Publication";"Publisher";"Image-URL-S";"Image-URL-M";"Image-URL-L"'][index].split(';')[1]
  temp_dict['Book-Title'] = removeLeadingAndEndingQuotes(temp_dict['Book-Title'])

  temp_dict['Book-Author'] = bx_books_df['ISBN;"Book-Title";"Book-Author";"Year-Of-Publication";"Publisher";"Image-URL-S";"Image-URL-M";"Image-URL-L"'][index].split(';')[2]
  temp_dict['Book-Author'] = removeLeadingAndEndingQuotes(temp_dict['Book-Author'])

  temp_dict['Year-Of-Publication'] = bx_books_df['ISBN;"Book-Title";"Book-Author";"Year-Of-Publication";"Publisher";"Image-URL-S";"Image-URL-M";"Image-URL-L"'][index].split(';')[3]
  temp_dict['Year-Of-Publication'] = removeLeadingAndEndingQuotes(temp_dict['Year-Of-Publication'])

  temp_dict['Publisher'] = bx_books_df['ISBN;"Book-Title";"Book-Author";"Year-Of-Publication";"Publisher";"Image-URL-S";"Image-URL-M";"Image-URL-L"'][index].split(';')[4]
  temp_dict['Publisher'] = removeLeadingAndEndingQuotes(temp_dict['Publisher'])

  temp_dict['Image-URL-S'] = bx_books_df['ISBN;"Book-Title";"Book-Author";"Year-Of-Publication";"Publisher";"Image-URL-S";"Image-URL-M";"Image-URL-L"'][index].split(';')[5]
  temp_dict['Image-URL-S'] = removeLeadingAndEndingQuotes(temp_dict['Image-URL-S'])

  temp_dict['Image-URL-M'] = bx_books_df['ISBN;"Book-Title";"Book-Author";"Year-Of-Publication";"Publisher";"Image-URL-S";"Image-URL-M";"Image-URL-L"'][index].split(';')[6]
  temp_dict['Image-URL-M'] = removeLeadingAndEndingQuotes(temp_dict['Image-URL-M'])

  temp_dict['Image-URL-L'] = bx_books_df['ISBN;"Book-Title";"Book-Author";"Year-Of-Publication";"Publisher";"Image-URL-S";"Image-URL-M";"Image-URL-L"'][index].split(';')[7]
  temp_dict['Image-URL-L'] = removeLeadingAndEndingQuotes(temp_dict['Image-URL-L'])

  temp_list.append(temp_dict)

books_df = pd.DataFrame(temp_list)           


```

```
list_alphabets = []
list_alphabets = [chr(x) for x in range(ord('A'), ord('Z') + 1)] 
```
